# The Caesar Cipher

A Caesar cipher is one of the simplest and most widely known encryption techniques used in cryptography. It is a type of substitution cipher in which each letter in the plaintext is replaced by a letter some fixed number of positions along the alphabet. For example, with a left shift of 3, D would be replaced by A, E would become B, and so on. The method is named after Julius Caesar, who used it in his private correspondence.

## Syneraio

This cipher was (probably) invented and used by Gaius Julius Caesar and his troops during the Gallic Wars. The idea is rather simple – every letter of the message is replaced by its nearest consequent (A becomes B, B becomes C, and so on). The only exception is Z, which becomes A.


```python
# Caesar cipher-encrypting message
text = input("Enter your message: ")
cipher = ''
for char in text:
    if not char.isalpha():
        continue
    char = char.upper()
    code = ord(char) + 1
    if code > ord('Z'):
        code = ord('A')
    cipher += chr(code)

print(cipher)
```

```python
# Caesar cipher - decrypting a message.
cipher = input('Enter your cryptogram: ')
text = ''
for char in cipher:
    if not char.isalpha():
        continue
    char = char.upper()
    code = ord(char) - 1
    if code < ord('A'):
        code = ord('Z')
    text += chr(code)

print(text)  
```

## CONCLUSION

The Caesar Cipher is one of the simplest and oldest encryption techniques. It works by shifting each letter in a message by a fixed number of positions in the alphabet, making the original text difficult to understand without knowing the shift value. Although it is not secure enough for modern communication, it introduces important concepts such as character manipulation, ASCII values, loops, and basic cryptography. Implementing the Caesar Cipher in Python helps in understanding string processing and serves as a foundation for learning more advanced encryption algorithms.
